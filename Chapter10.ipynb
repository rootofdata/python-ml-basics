```python
frame = pd.DataFrame({'data1' : np.random.randn(1000),
                    'data2' : np.random.randn(1000)})

quartiles = pd.cut(frame.data1, 4) #pd.cut : 동일 길이로 나누어서 범주 만들기
quartiles # -3.3~-1.574 , -1.574~-0.145, 0.145 ~ 1.863 , 1.863 ~ 3.582
```

```python
#cut에서 반환된 범주형객체는 바로 groupby로 넘길 수 있음 !
def get_stats(group):
    return {'min' : group.min(), 'max' : group.max(),
            'count' : group.count(), 'mean' : group.mean()}

grouped = frame.data2.groupby(quartiles)
grouped.apply(get_stats).unstack()
```

```python
suits = ['H', 'S', 'C', 'D'] #하트, 스페이드, 클럽, 다이아몬드
card_val = (list(range(1,11)) + [10]*3)*4
base_names = ['A'] + list(range(2,11)) + ['J', 'K', 'Q']
cards = []
for suit in ['H', 'S', 'C', 'D'] :
    cards.extend(str(num) + suit for num in base_names)

deck = pd.Series(card_val, index= cards)
deck
```

```python
def draw(deck, n = 5) : 
    return deck.sample(n)

draw(deck)
```

```python
get_suit = lambda card : card[-1] # 마지막 글자가 세트
deck.groupby(get_suit).apply(draw, n = 2)
# 각 세트별로 2장의 카드 무작위 추출
```

```python
deck.groupby(get_suit, group_keys = False).apply(draw, n = 2)
```

```python
spx_corr = lambda x: x.corrwith(x['SPX']) #'SPX'컬럼과 다른 컬럼의 상관관계를 계산
```

```python
rets = close_px.pct_change().dropna() # close_px의 퍼센트 변화율 계산 
get_year = lambda x: x.year # 각 datetime에서 연도 속성만 반환하는 함수

by_year = rets.groupby(get_year)
by_year.apply(spx_corr) # 연도벼 퍼센트 변화율
```

```python
#두 컬럼 간의 상관관계 계산
by_year.apply(lambda g: g['AAPL'].corr(g['MSFT']))
```

```python
import statsmodels.api as sm
def regress(data, yvar, xvars):
    Y = data[yvar]
    X = data[xvars]
    X['intercept'] = 1.
    result = sm.OLS(Y, X).fit()
    return result.params
```

```python
by_year.apply(regress, 'AAPL', ['SPX'])
```

```python
#피벗테이블 : 데이터 요약화 도구 -> 데이터를 키로 수집해서 어떤 키는 로우에, 어떤 키는 컬럼에 나열해서 데이터를 정렬
#pivot_table -> pandas 모듈의 최상위 함수
#마진 : 부분합을 추가할 수 있는 기능을 제공
tips.pivot_table(index = ['day', 'smoker']) #요일과 흡연자 집단에서 평균 (pivot_table의 기본 연산)
```
