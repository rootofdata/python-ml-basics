```python
a = int(input())
for i in range(a):
    re=0
    b = list(map(int,input().split()))   
    c = (sum(b)-b[0])/(len(b)-1)
    for j in len(b):
        if c>b[j+1]:
            re+=1
    result=re/len(b)
    print(result)    
    b=[]
    c=0
```
