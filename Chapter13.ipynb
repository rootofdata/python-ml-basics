```python
import statsmodels.api as sm
import statsmodels.formula.api as smf
```

```python
def dnorm(mean, variance, size = 1): 
    if isinstance(size, int):
        size = size,
    return mean + np.sqrt(variance)*np.random.randn(*size)

np.random.seed(12345)

N = 100
X = np.c_[dnorm(0, 0.4, size = N),
        dnorm(0, 0.6, size = N),
        dnorm(0, 0.2, size = N)]
eps = dnorm(0, 0.1, size =N)
beta = [0.1, 0.3, 0.5]

y = np.dot(X, beta) + eps
```

```python
help(dnorm) #dnorm : 정규분포 데이터 생성
```

```python
X[:5]
```

```python
y[:5]
```

```python
X_model = sm.add_constant(X) #선형 모델의 intercept을 기존 행렬에 더하기
X_model[:5]
```

```python
model = sm.OLS(y, X) # OLS : 최소제곱 / sm.OLS : 최소제곱 선형회귀
results = model.fit() #model.fit : 회귀 결과 객체를 반환
results.params
```

```python
print(results.summary())
```

```python
#특정 승객이 생존할 것인지 예측하는 것이 과제
#학습 데이터셋에 모델을 피팅하고 나서 테스트 데이터셋으로 검증
#결측치를 보완하기 위한 여러가지 방법 중 학습데이터 셋의 중간값으로 채워 넣자
impute_value = train['Age'].median()
train['Age'] = train['Age'].fillna(impute_value)
test['Age'] = test['Age'].fillna(impute_value)
```

```python
#IsFemale 컬럼을 추가해서 'Sex'컬럼을 인코딩
train['IsFemale'] = (train['Sex'] == 'female').astype(int)
test['IsFemale'] = (test['Sex'] == 'female').astype(int)
predictors = ['Pclass', 'IsFemale', 'Age']
X_train = train[predictors].values
X_test = test[predictors].values
y_train = train['Survived'].values
X_train[:5]
```

```python
#scikit-learn의 logisticRegression 모델을 이용해서 model 인스턴스를 생성
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
```

```python
#model의 fit 메서드를 이용해서 이 모델을 학습 데이터에 피팅
model.fit(X_train, y_train)
```

```python
y_predictor = model.predict(X_test) #model.predict : 테스트 데이터셋으로 예측
y_predictor[:10]
```

```python
#학습 데이터에 오버피팅되는 것을 피할 수 있도록 교차검증 같은 기법을 활용 => 새로운 데이터에 대해 좀 더 견고하거나 예측 성능을 높여줌
from sklearn.linear_model import LogisticRegressionCV
model_cv = LogisticRegressionCV(10)
model_cv.fit(X_train, y_train)
```
