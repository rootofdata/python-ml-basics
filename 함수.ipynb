```python
def d(n):
    ans=n
    for i in range(0,len(str(n))):
        ans += int(n/10 **i)%10
    return ans
self_number=list()

for i in range(1,10001):
    self_number.append(d(i))
for i in range(1,10001):
    if i not in self_number:
        print(i)
```

```python
n= int(input())
result=[]
for i in range(1,n+1):
    li=[]
    if i<100:
        result.append(i)
    else:
        for j in str(i):
            li.append(int(j))
        a=li[0]-li[1]
        b=li[1]-li[2]
        if a==b:
            result.append(i)
print(len(result))
```
